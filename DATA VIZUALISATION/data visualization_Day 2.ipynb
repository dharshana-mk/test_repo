```python
###Joint plots##
sns.set_style("dark")
data
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data =data,hue = 'time')
```

```python
###Regplot --> performs simple linear regression and 
sns.regplot(x = 'tip',y ='total_bill',data=data)
```

```python
sns.regplot(x = 'tip',y ='total_bill',data=data,x_jitter = 1)
```

```python
sns.regplot(x = 'size',y ='total_bill',data=data,x_jitter = 0.5)
```

```python
sns.regplot(x = 'size',y ='total_bill',data=data)
```

```python
##lmplot --> performs simple linear regression and 
sns.lmplot(x = 'total_bill',y ='tip',hue = 'day', data=data, height = 3,col='day')
```

```python
##lmplot --> performs simple linear regression and 
sns.lmplot(x = 'total_bill',y ='tip',col='day',hue = 'day', data=data,col_wrap = 2, height = 3)
```

```python
sns.boxplot(x='day',y='total_bill',data = data)
```
