```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataframe

df = pd.read_excel("Mall_Customers_data.xlsx")
df.head()
```

1. What is the average annual income of female customers who have spending score of less than 65? 

```python
spend_score = df[(df["Spending Score (1-100)"]<65) & (df["Gender"]=="Female")]
spend_score
```

```python
print("The avg annual income of female customers with spending score < 65: ", spend_score["Annual Income (k$)"].mean())
```

2. Explore the relationship (if any) between gender, annual income as well as spending score. What are the conclusions that may be drawn?

```python
df.pivot_table(index=["Gender"], values=["Annual Income (k$)", "Spending Score (1-100)"], aggfunc="mean")
```

Females tend to spend more on an average than males, even with the former's average annual income falling a little short than the latter's.

3. Explore the correlation (if any) between annual income and spending score with respect to age group.

```python
sns.scatterplot(data=df, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Age Group")
```

4. Determine the highest spending score of an adult male.

```python
adult_male_df = df[(df["Gender"]=="Male") & (df["Age Group"]=="Adult")]
adult_male_df
```

```python
print("The highest spending score of an adult male: ", max(adult_male_df["Spending Score (1-100)"]))
```
